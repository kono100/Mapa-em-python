Mapa em New York

```python
import folium
from IPython.display import display

map_center = [40.7128, -74.0060]
mymap = folium.Map(location=map_center, zoom_start=12)

folium.Marker(
    [40.7128, -74.0060],
    popup="New York",
    icon=folium.Icon(color="blue", icon="info-sign")
).add_to(mymap)

display(mymap)
```

Minha localização \ Cidade

```python
import folium
import geocoder
from IPython.display import display

# Obtendo a localização atual pelo IP
g = geocoder.ip('me')  # Obtém a localização com base no IP
latitude, longitude = g.latlng  # Extrai latitude e longitude

# Criação do mapa centrado na sua localização
map_center = [latitude, longitude]
mymap = folium.Map(location=map_center, zoom_start=12)

# Adicionando um marcador na sua localização
folium.Marker(
    [latitude, longitude],
    popup="Você está aqui!",
    icon=folium.Icon(color="green", icon="info-sign")
).add_to(mymap)

# Exibindo o mapa
display(mymap)
```

TRACEJADO - RANDON

```python
import folium
import random

# Ponto inicial para o mapa (latitude e longitude base)
latitude = -23.55052  # Latitude inicial (São Paulo, como exemplo)
longitude = -46.633308  # Longitude inicial

# Criação do mapa centrado no ponto inicial
map_center = [latitude, longitude]
mymap = folium.Map(location=map_center, zoom_start=15)

# Lista para armazenar os pontos visitados
points = [[latitude, longitude]]

# Gerar 10 pontos aleatórios ao redor da posição inicial
for _ in range(10):
    # Gerar deslocamentos pequenos em latitude e longitude
    lat_offset = random.uniform(-0.01, 0.01)
    lon_offset = random.uniform(-0.01, 0.01)
    
    # Atualizar latitude e longitude
    latitude += lat_offset
    longitude += lon_offset
    
    # Adicionar o novo ponto à lista
    points.append([latitude, longitude])
    
    # Adicionar um marcador no novo ponto
    folium.Marker(
        [latitude, longitude],
        popup=f"Localização: ({latitude:.5f}, {longitude:.5f})",
        icon=folium.Icon(color="blue", icon="info-sign")
    ).add_to(mymap)

# Adicionar uma linha conectando os pontos
folium.PolyLine(points, color="red", weight=2.5, opacity=1).add_to(mymap)

# Salvar o mapa em um arquivo HTML
mymap.save("random_points_map.html")
print("Mapa criado! Abra o arquivo 'random_points_map.html' no navegador para visualizar.")
```

TRACEJADO - RANDON em REAL-TIME

```python
import folium
import random
import time

# Ponto inicial para o mapa (latitude e longitude base)
latitude = -23.55052  # Latitude inicial (São Paulo, como exemplo)
longitude = -46.633308  # Longitude inicial

# Criação do mapa centrado no ponto inicial
map_center = [latitude, longitude]
mymap = folium.Map(location=map_center, zoom_start=15)

# Lista para armazenar os pontos visitados
points = [[latitude, longitude]]

# Função para salvar o mapa atualizado
def save_map():
    mymap.save("real_time_map.html")
    print("Mapa salvo!")

# Inicializa o mapa com o ponto inicial
folium.Marker([latitude, longitude], popup="Localização Inicial", icon=folium.Icon(color="blue", icon="info-sign")).add_to(mymap)
save_map()  # Salva o mapa inicial

# Gerar 10 pontos aleatórios ao redor da posição inicial
for i in range(10):
    # Gerar deslocamentos pequenos em latitude e longitude
    lat_offset = random.uniform(-0.01, 0.01)
    lon_offset = random.uniform(-0.01, 0.01)
    
    # Atualizar latitude e longitude
    latitude += lat_offset
    longitude += lon_offset
    
    # Adicionar o novo ponto à lista de pontos
    points.append([latitude, longitude])
    
    # Adicionar um novo marcador no novo ponto
    folium.Marker(
        [latitude, longitude],
        popup=f"Localização: ({latitude:.5f}, {longitude:.5f})",
        icon=folium.Icon(color="blue", icon="info-sign")
    ).add_to(mymap)

    # Adicionar uma linha conectando os pontos até o momento
    folium.PolyLine(points, color="red", weight=2.5, opacity=1).add_to(mymap)

    # Salvar o mapa atualizado
    save_map()

    # Imprimir progresso no console
    print(f"Ponto {i + 1} adicionado: ({latitude:.5f}, {longitude:.5f})")

    # Esperar 3 segundos antes de gerar o próximo ponto
    time.sleep(3)

# Mensagem final
print("Rastreamento finalizado! Abra o arquivo 'real_time_map.html' no navegador para visualizar o trajeto.")
```
